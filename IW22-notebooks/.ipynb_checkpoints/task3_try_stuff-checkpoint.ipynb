```python
#for points with 2 dimensions volume collapses to Area
def voronoi_volumes(points):
    v = Voronoi(points)
    vol = np.zeros(v.npoints)
    for i, reg_num in enumerate(v.point_region):
        indices = v.regions[reg_num]
        if -1 in indices: # some regions can be opened
            vol[i] = np.inf
        else:
            vol[i] = ConvexHull(v.vertices[indices]).volume
    return vol
```

```python
from shapely.geometry import Point
# Turn points into list of x,y shapely points 
wow19_locations = [Point(xy) for xy in points19]
wow19_locations
# Create geodataframe using the points
wow19_locations = gpd.GeoDataFrame(wow19_locations, 
                                  columns=['geometry'],
                                  crs="EPSG:4326")
wow19_locations_RD = wow19_locations.to_crs("EPSG:28992")

knmi_locations = [Point(xy) for xy in knmiPoints]

# Create geodataframe using the points
knmi_locations = gpd.GeoDataFrame(knmi_locations, 
                                  columns=['geometry'],
                                  crs="EPSG:4326")
knmi_locations_RD = knmi_locations.to_crs("EPSG:28992")


```

```python
wow_19_list_RD = [(x,y) for x,y in zip(wow19_locations_RD.geometry.x , wow19_locations_RD.geometry.y)]
knmi_st_list_RD = [(x,y) for x,y in zip(knmi_locations_RD.geometry.x , knmi_locations_RD.geometry.y)]
```

```python
areas = voronoi_volumes(wow_19_list_RD)
areasKNMI = voronoi_volumes(knmi_st_list_RD)
```

```python
len(areas)
```

```python
def filterAndPlotArea(vorArea,title):
    areas_fil = vorArea[np.isfinite(vorArea)]
    areas_fil = np.sort(areas_fil)/1e6
    fig, ax = plt.subplots(figsize=(50,20))
    ax.bar(np.arange(0,len(areas_fil)),areas_fil)
    ax.set_yscale('log')
    matplotlib.rcParams.update({'font.size': 22})
    plt.title(title,fontsize = 40)
    plt.xlabel("polygon #",fontsize = 40)
    plt.ylabel("surface area (sq. km)",fontsize = 40)

    
    plt.show()

#plt.hist(areas)
```

```python
filterAndPlotArea(areas,"WOW voronoi surface area")
filterAndPlotArea(areasKNMI,"KNMI stations voronoi surface area")
```

```python
def filterAndPlotAreaTogether(vorAreaWOW, vorAreaKNMI, title):
    areas_fil = vorAreaWOW[np.isfinite(vorAreaWOW)]
    areas_fil = np.sort(areas_fil)/1e6
    areas_fil2 = vorAreaKNMI[np.isfinite(vorAreaKNMI)]
    areas_fil2 = np.sort(areas_fil2)/1e6
    fig, ax = plt.subplots(figsize=(40,30))
    bl2= ax.bar(np.arange(0,len(areas_fil2)),areas_fil2,color='r')
    bl1= ax.bar(np.arange(len(areas_fil2),len(areas_fil2)+len(areas_fil)),areas_fil,color='b')
    ax.set_yscale('log')
    matplotlib.rcParams.update({'font.size': 22})
    plt.title(title,fontsize = 40)
    plt.xlabel("polygon #",fontsize = 40)
    plt.ylabel("voronoi cell surface area (sq. km)",fontsize = 40)
    plt.legend((bl2, bl1), ('KNMI', 'WOW'),prop={'size':55})
    #bl1.set_color('r')
    #bl2.set_color('b')

    plt.savefig("task3_voronoi_regions_surface_area_KNMIvsWOW.png")
    plt.show()
```

```python
filterAndPlotAreaTogether(areas,areasKNMI,"KNMI vs. WOW voronoi surface areas")
```
