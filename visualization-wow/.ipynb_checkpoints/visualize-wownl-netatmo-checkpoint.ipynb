# Visualizing WOW-NL and Netatmo networks

This notebook makes a simple plot to visualize WOW-NL and Netatmo networks side by side. Both data collections were processed locally to obtain a summary raster file showing the number of Personal Weather Stations (PWS) per 1km pixel in the Netherlands. 

**NOTE:** Netatmo data was acquired via a EUMETNET project in 2022 and was processed to create the summary raster file. The data agreement signed with EUMETNET/Netatmo states that Netatmo data can be used for research, but must not be freely distributed (at least without consultation). 

First we import the required libraries

```python
import cartopy
import matplotlib
import xarray as xr
import cartopy.crs as ccrs
import matplotlib.pyplot as plt
import shutup; shutup.please()
import cartopy.feature as cfeature
import geopandas as gpd
from cartopy.io.shapereader import Reader
```

Setting up paths and global variables

```python
path_wow = r"./geodata/raster/WOW_and_Netatmo/Clipped_PiP_WOW_RDNew_rasterized.nc"
path_netatmo = r"./geodata/raster/WOW_and_Netatmo/Clipped_PiP_Netatmo_RDNew_rasterized.nc"
path_prov = r"./geodata/vector/NL_provinces_gpkg_RDNew/NL_Provinces_RDNew.gpkg"

projection = ccrs.epsg(28992)
xmin, xmax, ymin, ymax = [130000, 200000, 357000, 427000]
gdf_prov = gpd.read_file(path_prov, layer='GRS_1000_PROV_NL_V')
# prov_geometries = list(Reader(path_prov, bbox=[xmin, ymin, xmax, ymax]).geometries())

resolution = "50m"
ocean = cartopy.feature.NaturalEarthFeature('physical', 'ocean', scale=resolution, edgecolor='none', facecolor=cfeature.COLORS['water'])
boundaries = cartopy.feature.NaturalEarthFeature(category='cultural', name='admin_0_boundary_lines_land', scale=resolution, facecolor='none', alpha=0.7)
land = cartopy.feature.NaturalEarthFeature('physical', 'land', scale=resolution, edgecolor='k', facecolor=cfeature.COLORS['land'])
```

Now we map the two datasets side by side.

**NOTE:** Visualization takes some time because the rasters are 64MB for unknown reasons! :-) 

```python
ds_wow = xr.open_dataset(path_wow, decode_times=True, use_cftime=True)
ds_net = xr.open_dataset(path_netatmo, decode_times=True, use_cftime=True)

mycmap2 = matplotlib.colors.ListedColormap(['white', 'black'])
mycmap3 = matplotlib.colors.ListedColormap(['white', 'black', 'orange'])
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(24, 12), subplot_kw={'projection': projection})

ax[0].add_feature(ocean, zorder=0)
ax[0].add_feature(land, zorder=0)
ax[0].add_feature(boundaries, zorder=1)
mask_wow = xr.where(ds_wow.Band1 > 1, 1, ds_wow.Band1)
mask_wow.plot(ax=ax[0], cmap=mycmap2, zorder=2, cbar_kwargs={"orientation": "horizontal"})
gdf_prov.plot(ax=ax[0], facecolor="None", edgecolor='gray', zorder=3)
ax[0].set_title("WOW-NL network (2023)", size=20)

ax[1].add_feature(ocean, zorder=0)
ax[1].add_feature(land, zorder=0)
ax[1].add_feature(boundaries, zorder=1)
mask_net = xr.where(ds_net.Band1 > 1, 1, ds_net.Band1)
mask_net.plot(ax=ax[1], cmap=mycmap2, zorder=2, cbar_kwargs={"orientation": "horizontal"})
gdf_prov.plot(ax=ax[1], facecolor="None", edgecolor='gray', zorder=3)
ax[1].set_title("Netatmo network in the Netherlands (2020)", size=20)

ax[2].add_feature(ocean, zorder=0)
ax[2].add_feature(land, zorder=0)
ax[2].add_feature(boundaries, zorder=1)
mask_both =  mask_net + mask_wow
mask_both.plot(ax=ax[2], cmap=plt.cm.viridis, zorder=2, cbar_kwargs={"orientation": "horizontal"})
gdf_prov.plot(ax=ax[2], facecolor="None", edgecolor='gray', zorder=3)
ax[2].set_title("Coverage of WOW-NL + Netatmo", size=20)
```
