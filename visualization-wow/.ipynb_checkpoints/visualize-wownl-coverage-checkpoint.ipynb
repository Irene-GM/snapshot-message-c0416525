### Visualizing WOW-NL monthly coverage (October 2023)

Here a simple notebook counting the number of WOW-NL observations per grid cell. Note in the `geodata/vector` folder there is a `geopackage` file (`.gpkg`) containing a vector representation of the Netherlands at 1km, 5km, and 10km of spatial resolution. The coordinate reference system of this dataset is `EPSG:28992` (aka `RD_New`). To change the spatial aggregation, just change the `layer` parameter at the time of reading the `path_grid` file and re-run the notebook again (`Run > Restart kernel and run all cells`). Since `geopandas` is not the fastest library to compute spatial aggregations, it might take several minutes to run the visualization. 

<span style="color:green">**Quick notes**:</span> 
- The WOW-NL network yielded around 8 million observations in Oct 2023, but part of them fall in BES, Belgium and Germany. They are filtered using the grid cells pertaining to the Netherlands.
- The visualized dataset does not contain any of the observations from KNMI stations, since we filtered them when querying the WOW Livestream by `IsOfficial=False`.

#### Imports and global variables

```python
import cartopy
import matplotlib
import numpy as np
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
from shapely.geometry import box
import cartopy.crs as ccrs
import cartopy.feature as cfeature

pd.options.display.max_columns = 500
pd.options.display.width = 0

projection = ccrs.epsg(28992)
extent = (3, 8, 50, 54)

resolution = "50m"

boundaries = cartopy.feature.NaturalEarthFeature(category='cultural', name='admin_0_boundary_lines_land', scale=resolution, facecolor='none', alpha=0.7)
land = cartopy.feature.NaturalEarthFeature('physical', 'land', scale=resolution, edgecolor='k', facecolor=cfeature.COLORS['land_alt1'])
ocean = cartopy.feature.NaturalEarthFeature('physical', 'ocean', scale=resolution, edgecolor='none', facecolor=cfeature.COLORS['water'])
lakes = cartopy.feature.NaturalEarthFeature('physical', 'lakes', scale=resolution, edgecolor='k', facecolor=cfeature.COLORS['water'])
```

#### Reading the grid and the monthly observations

```python
path_obs = r"./data-common/processed_wow/WOW-NL_Oct23.csv"
path_grid = r"./geodata/vector/NL_RDNew_grid_clip.gpkg"

df = pd.read_csv(path_obs, sep=";", header=0, index_col=["rowid"])
gdf_obs = gpd.GeoDataFrame(df, geometry=gpd.points_from_xy(df.Longitude, df.Latitude), crs="EPSG:4326")
gdf_obs = gdf_obs.to_crs(epsg=28992)
gdf_grid = gpd.read_file(path_grid, layer='NL_RDNew_5km_grid_clip') # spatial resolutions available: 1km, 5km, 10km

print("Num. obs. in Oct23: ", gdf_obs.shape)
print("Num. cells in grid: ", gdf_grid.shape)
```

#### Counting observations per grid cell (the slow & lazy part!)
<span style="color:green">**Note:**</span> If you see a map in the next jupyter cell, you might skip re-running this part

```python
l = []

for idx, row in gdf_grid.iterrows():    
    bbox_cell = box(row["left"], row["bottom"], row["right"], row["top"])
    gdf_obs_cell = gpd.clip(gdf_obs, mask=bbox_cell)
    l.append(gdf_obs_cell.shape[0])

gdf_grid["count"] = np.array(l)
print("Min/max values found: {0}/{1}".format(np.amin(l), np.amax(l)))
```

#### Plotting the map showing the observations per grid cell
<span style="color:green">**Note:**</span> to remove the log-normalization of the color bar, delete the parameter `norm` from the `.plot()` function.

```python
log_norm = matplotlib.colors.LogNorm(vmin=1, vmax=gdf_grid["count"].max())

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(15, 12), subplot_kw={'projection': projection})
ax.set_title("Number of observations per grid cell", size=18)
ax.gridlines(linestyle='-', linewidth=0.5)
ax.add_feature(ocean, zorder=1)
ax.coastlines(resolution=resolution, linewidth=1.5, zorder=2)
ax.add_feature(land, zorder=3)
ax.add_feature(lakes, linewidth=1.5, zorder=4)
gdf_grid.plot(column="count", ax=ax, cmap=plt.cm.plasma, zorder=5, legend=True, norm=log_norm, legend_kwds={"label":"obs. per cell (log scale)"})
ax.add_feature(boundaries, linestyle='-', linewidth=1.5, zorder=6)
```
